```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (8, 3)
```

```python
extended_1 = pd.read_table("extended_1.XYZ", sep = " ", header = None)
unscented_1 = pd.read_table("unscented_1.XYZ", sep = " ", header = None)
```

```python
extended_1.head()
```

```python
unscented_1.head()
```

```python
plt.plot([i for i in range(extended_1.shape[1] - 1)], [max(m.log(i), 1) for i in extended_1.iloc[0, :-1]], label="Расширенный")
plt.plot([i for i in range(unscented_1.shape[1] - 1)], [max(m.log(i), 1) for i in unscented_1.iloc[0, :-1]], label="Ансцентный")
plt.xlabel("Итерации")
plt.ylabel("Невязка (логарифм)")
plt.legend()
plt.savefig('residual_1.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
extended_var = pd.read_table("extended_5_var.XYZ", sep = " ", header = None)
unscented_var = pd.read_table("unscented_5_var.XYZ", sep = " ", header = None)
```

```python
plt.rcParams['figure.figsize'] = (4, 3)
```

```python
tmp = np.array(extended_var.iloc[:, :-1])
ext_var = np.matmul(tmp, tmp.T)
sns.heatmap(ext_var)
plt.title("Ковариация параметров")
plt.savefig('extended_var.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
tmp = np.array(unscented_var.iloc[:, :-1])
unc_var = np.matmul(tmp, tmp.T)
sns.heatmap(unc_var)
plt.title("Ковариация параметров")
plt.savefig('extended_var.png', dpi=600, bbox_inches='tight')
plt.show()
```
